```python
with open('day3_input.txt', 'r') as f:
    data = f.readlines()
```

```python
data = [x.rstrip() for x in data]
data
```

```python
index = 0
trees = 0
for row in data:
    while index >= len(row):
        row *= 2
    if row[index] == '#':
        trees += 1
    index += 3
trees
```

```python
## part 2
tree_counter = trees
slope = {0: [1,1], 1: [5,1], 2: [7,1], 3: [1,2]}
for key in slope.keys():
    index = 0
    trees = 0
    row_count = 0
    for row in data:
        if row_count % slope[key][1] == 0:
            while index >= len(row):
                row *= 2
        
            if row[index] == '#':
                trees += 1
            index += slope[key][0]
        row_count += 1
    tree_counter *= trees
tree_counter
    
```
