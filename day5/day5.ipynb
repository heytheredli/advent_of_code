```python
with open('day5_input.txt', 'r') as f:
    data = f.readlines()
```

```python
data = [x.rstrip() for x in data]
```

```python
def find_row(bpass):
    row = 128
    row_bound = [0, 127]
    for letter in bpass:
        row /= 2
        if letter == 'F':
            row_bound = [row_bound[0], row_bound[0] + row - 1]
        elif letter == 'B':
            row_bound = [row_bound[0] + row, row_bound[1]]
    return row_bound[0]

def find_col(bpass):
    col = 8
    col_bound = [0, 7]
    for letter in bpass:
        col /= 2
        if letter == 'L':
            col_bound = [col_bound[0], col_bound[0] + col - 1]
        elif letter == 'R':
            col_bound = [col_bound[0] + col, col_bound[1]]
    return col_bound[0]

def seat_id(bpass):
    bpass_row = bpass[:7]
    bpass_col = bpass[7:]
    return (find_row(bpass_row) * 8) + find_col(bpass_col)
```

```python
max_seat_id = 0
for bpass in data:
    cur_seat_id = seat_id(bpass)
    if cur_seat_id > max_seat_id:
        max_seat_id = cur_seat_id
max_seat_id
```

```python
## Part 2

seat_ids = []
for bpass in data:
    seat_ids.append(seat_id(bpass))
for i in range(1, 127):
    for j in range(8):
        cur_id = i * 8 + j
        if cur_id not in seat_ids:
            print(cur_id)
```
